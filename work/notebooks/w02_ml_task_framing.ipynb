## 1. My lane as an ML task (type)

*Classification, clustering, ranking, or scoring — which one, and why?*

```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.
# My Lane as an ML Task

**Lane:** Refresh / Content Opportunity Scoring

**ML Task Type:** Ranking / Scoring

The goal is to rank webpages according to how urgently they should be reviewed for content refresh. The model assigns a refresh score so that the content team can prioritize the most important pages first.
```

## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*

```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.
# Target or Proxy

The target is whether a webpage is likely to require a content refresh.

The starter dataset uses the derived label `is_declining_label`, which is based on `trend_direction`. This label can be used as the target for learning which pages should be prioritized.
```

## 3. Success metric

*One metric you can defend. What number means 'good'?*

```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.
# Success Metric

The main success metric is Precision@50.

A higher Precision@50 means more of the top-ranked webpages truly need review, allowing the content team to focus on the most important pages.
```

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.
# Why ML Beats a Fixed Rule

A fixed rule uses only one or two conditions and may miss important patterns.

Machine learning can combine many features at the same time and learn relationships from historical data.

This allows the model to produce better recommendations than a simple rule-based approach.
```
